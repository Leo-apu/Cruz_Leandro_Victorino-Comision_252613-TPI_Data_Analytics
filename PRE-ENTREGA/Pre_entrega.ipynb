## Script básico
Actividades 2 y 3

En esta etapa se desarrolló un script para calcular el total de ventas mensuales utilizando distintas técnicas de programación en Python.

Se optó por almacenar los datos en una lista de diccionarios, ya que permite organizar la información de cada producto (nombre, precio, cantidad) de forma estructurada.

Se implementaron dos métodos de cálculo del total de ventas:

- List comprehension: por su claridad y eficiencia.
- Función map() con lambda: por su estilo funcional y expresivo.

Finalmente, se muestran los resultados individuales por producto y el total general.

```python
# Generamos una lista de diccionarios que contenga algunos datos
# Cada elemento representa un producto vendido durante el mes.
ventas = [
    {"producto": "Auriculares Bluetooth", "precio": 45.0, "cantidad": 20},
    {"producto": "Tablet 10 pulgadas", "precio": 320.0, "cantidad": 7},
    {"producto": "Cargador USB-C", "precio": 18.0, "cantidad": 25},
    {"producto": "Parlante portátil", "precio": 65.0, "cantidad": 12},
    {"producto": "Smartwatch deportivo", "precio": 210.0, "cantidad": 9},
    {"producto": "Cámara web HD", "precio": 85.0, "cantidad": 6},
]
```

```python
# Calculamos las ventas acumuladas - map
ventas_acumuladas_map = sum(map(lambda v: v["precio"] * v["cantidad"], ventas))
ventas_acumuladas_map
```

```python
# Calculamos las ventas acumuladas - listas por comprensión
venta_acumuladas = sum([venta["precio"] * venta["cantidad"] for venta in ventas])
venta_acumuladas
```

```python
# ---------------------------------------------------------
# DETALLE DE VENTAS POR PRODUCTO
# ---------------------------------------------------------
print("Detalle de ventas por producto:")
for venta in ventas:
    total_producto = venta["precio"] * venta["cantidad"]
    print(f"- {venta['producto']}: {venta['cantidad']} unidades → ${total_producto:.2f}")

# ---------------------------------------------------------
# REFLEXIÓN:
# Se utilizó una lista de diccionarios porque permite estructurar
# los datos de manera ordenada, relacionando cada producto con su precio
# y cantidad. Además, facilita aplicar distintos métodos de cálculo,
# tanto imperativos como funcionales.
```

```python
# Transformación:
# Calculamos el valor de cada venta = precio * cantidad
# Agregamos una nueva columnas
df_ventas["valor_venta"] = df_ventas["precio"] * df_ventas["cantidad"]
```

## Integración de datos
Actividad 4 - Combinar los sets de datos de ventas y marketing para obtener una visión más amplia de las tendencias.

```python
# Mostrar el encabezado del datasets ventas para verificar que columna de repite en el dataset marketing
df_ventas.head(2)
```

```python
# Mostrar la informacion del datasets de marketing para comparar con los encabezados del datasets ventas para elegir la columna de union
df_marketing.info()
```

```python
# Antes de hacer la union debemos hacer cambios en la tabla de marketing , ya que si hacemos ahora el merge vamos encontrar inconcistencias en
# el datasets resultante. Para ellos unificamos la informacion del datasets marketing haciendo un promedio de los costos, cambiando el nombre de la columna
# id_campaha por num_campaña y con esto vamos a tener datos correctos para su analisis.

df_marketing_resumen = (
    df_marketing
    .groupby("producto", as_index=False)
    .agg({
        "costo": "mean",         # costo promedio de las campañas
        "id_campanha": "count",  # cantidad total de campañas
        "fecha_inicio": "min",   # fecha más temprana de inicio
        "fecha_fin": "max"       # fecha más tardía de finalización
    })
    .rename(columns={
        "id_campanha": "num_campañas"
    })
)
```

```python
# Unir los datasets ventas y marketing por columna 'producto'
ventas_marketing = pd.merge(df_ventas, df_marketing_resumen, on="producto", how="left")
```

```python
# Verificar coincidencias (muestra)
ventas_marketing.head()
```

```python
# Mostrar informacion general de la nueva tabla generada "ventas_marketing"
ventas_marketing.info()
```

```python
# Mostrar operaciones realizadas en la tabla "ventas_marketing" para usar a un futuro en datos estadisticos
ventas_marketing.describe()
```

```python
# Verificar si hay campos nulos
ventas_marketing.isnull().sum()
```

```python
# Verificar si hay campos duplicados
ventas_marketing.duplicated().sum()
```

```python
# OTRA OPCION PARA UNIFICAR DATASETS VERIFICANDO SI EXISTE EL CAMPO EN AMBOS.
# Unifico ventas con marketing por columna 'producto' verificando si existe
if 'producto' in df_ventas.columns and 'producto' in df_marketing.columns:
    # Resumir información de marketing por producto
    df_marketing_resumen = (
        df_marketing
        .groupby("producto", as_index=False)
        .agg({
            "costo": "mean",         # costo promedio de las campañas
            "id_campanha": "count",  # cantidad total de campañas
            "fecha_inicio": "min",   # fecha más temprana de inicio
            "fecha_fin": "max"       # fecha más tardía de finalización
        })
        .rename(columns={
            "id_campanha": "num_campañas"
        })
    )

    # Unificar datasets de ventas y marketing resumido
    df_combinado = pd.merge(df_ventas, df_marketing_resumen, on='producto', how='left')

    print("Dataset combinado correctamente sin duplicados.")
    print("Tamaño del dataset combinado:", df_combinado.shape)
    display(df_combinado.head())

else:
    print("No se encontró la columna 'producto' en ambos datasets para realizar la unión.")
```

### Observaciones



- Se realizó la integración entre los datasets de ventas y marketing utilizando la columna 'producto' como clave común.
- Se conservaron todas las ventas, aun si no existía información asociada en marketing (LEFT JOIN).
- La tabla integrada servirá para analizar la relación entre campañas de marketing y el rendimiento en ventas.
- En etapas posteriores, se pueden aplicar filtros o agrupaciones adicionales para detectar correlaciones o patrones.

### Analizamos el impacto de la campaña

```python
# Nos quedamos solo con los registros de ventas durante la campaña de marketing
# Seleccionamos solo los campos que nos interesan: "producto", "valor_venta", "costo"
ventas_marketing_impacto = ventas_marketing[["categoria", "valor_venta", "costo"]]
```

```python
# Agrupamos por producto y hacemos una agregación por "valor de venta" y "costo"
ventas_agrupadas = ventas_marketing_impacto.groupby("categoria")[["valor_venta","costo"]].sum().reset_index()
```

```python
# Generamos una nueva columna con la diferencia entre lo vendido y lo invertido en la campaña
ventas_agrupadas["ganancia_neta"] = ventas_agrupadas["valor_venta"] - ventas_agrupadas["costo"]
```

```python
# Mostramos el resultado final
ventas_agrupadas
```

```python
ventas_agrupadas.plot(kind='bar', x='categoria', y='ganancia_neta', title='Ganancia Neta por Categoría')
```
