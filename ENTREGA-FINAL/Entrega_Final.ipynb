## Script básico
Actividades 2 y 3

En esta etapa se desarrolló un script para calcular el total de ventas mensuales utilizando distintas técnicas de programación en Python.

Se optó por almacenar los datos en una lista de diccionarios, ya que permite organizar la información de cada producto (nombre, precio, cantidad) de forma estructurada.

Se implementaron dos métodos de cálculo del total de ventas:

- List comprehension: por su claridad y eficiencia.
- Función map() con lambda: por su estilo funcional y expresivo.

Finalmente, se muestran los resultados individuales por producto y el total general.

```python
# Generamos una lista de diccionarios que contenga algunos datos
# Cada elemento representa un producto vendido durante el mes.
ventas = [
    {"producto": "Auriculares Bluetooth", "precio": 45.0, "cantidad": 20},
    {"producto": "Tablet 10 pulgadas", "precio": 320.0, "cantidad": 7},
    {"producto": "Cargador USB-C", "precio": 18.0, "cantidad": 25},
    {"producto": "Parlante portátil", "precio": 65.0, "cantidad": 12},
    {"producto": "Smartwatch deportivo", "precio": 210.0, "cantidad": 9},
    {"producto": "Cámara web HD", "precio": 85.0, "cantidad": 6},
]
```

```python
# Calculamos las ventas acumuladas - map
ventas_acumuladas_map = sum(map(lambda v: v["precio"] * v["cantidad"], ventas))
ventas_acumuladas_map
```

```python
# Calculamos las ventas acumuladas - listas por comprensión
venta_acumuladas = sum([venta["precio"] * venta["cantidad"] for venta in ventas])
venta_acumuladas
```

```python
# ---------------------------------------------------------
# DETALLE DE VENTAS POR PRODUCTO
# ---------------------------------------------------------
print("Detalle de ventas por producto:")
for venta in ventas:
    total_producto = venta["precio"] * venta["cantidad"]
    print(f"- {venta['producto']}: {venta['cantidad']} unidades → ${total_producto:.2f}")

# ---------------------------------------------------------
# REFLEXIÓN:
# Se utilizó una lista de diccionarios porque permite estructurar
# los datos de manera ordenada, relacionando cada producto con su precio
# y cantidad. Además, facilita aplicar distintos métodos de cálculo,
# tanto imperativos como funcionales.
```

## Integración de datos
Actividad 4 - Combinar los sets de datos de ventas y marketing para obtener una visión más amplia de las tendencias.

```python
# Mostrar el encabezado del datasets ventas para verificar que columna de repite en el dataset marketing
df_ventas.head(2)
```

```python
# Mostrar la informacion del datasets de marketing para comparar con los encabezados del datasets ventas para elegir la columna de union
df_marketing.info()
```

```python
# Antes de hacer la union debemos hacer cambios en la tabla de marketing , ya que si hacemos ahora el merge vamos encontrar inconcistencias en
# el datasets resultante. Para ellos unificamos la informacion del datasets marketing haciendo un promedio de los costos, cambiando el nombre de la columna
# id_campaha por num_campaña y con esto vamos a tener datos correctos para su analisis.

df_marketing_resumen = (
    df_marketing
    .groupby("producto", as_index=False)
    .agg({
        "costo": "mean",         # costo promedio de las campañas
        "id_campanha": "count",  # cantidad total de campañas
        "fecha_inicio": "min",   # fecha más temprana de inicio
        "fecha_fin": "max"       # fecha más tardía de finalización
    })
    .rename(columns={
        "id_campanha": "num_campañas"
    })
)
```

```python
# Unir los datasets ventas y marketing por columna 'producto'
ventas_marketing = pd.merge(df_ventas, df_marketing_resumen, on="producto", how="left")
```

```python
# Verificar coincidencias (muestra)
ventas_marketing.head()
```

```python
# Verificar si hay campos nulos
ventas_marketing.isnull().sum()
```

```python
# Verificar si hay campos duplicados
ventas_marketing.duplicated().sum()
```

```python
# OTRA OPCION PARA UNIFICAR DATASETS VERIFICANDO SI EXISTE EL CAMPO EN AMBOS.
# Unifico ventas con marketing por columna 'producto' verificando si existe
if 'producto' in df_ventas.columns and 'producto' in df_marketing.columns:
    # Resumir información de marketing por producto
    df_marketing_resumen = (
        df_marketing
        .groupby("producto", as_index=False)
        .agg({
            "costo": "mean",         # costo promedio de las campañas
            "id_campanha": "count",  # cantidad total de campañas
            "fecha_inicio": "min",   # fecha más temprana de inicio
            "fecha_fin": "max"       # fecha más tardía de finalización
        })
        .rename(columns={
            "id_campanha": "num_campañas"
        })
    )

    # Unificar datasets de ventas y marketing resumido
    df_combinado = pd.merge(df_ventas, df_marketing_resumen, on='producto', how='left')

    print("Dataset combinado correctamente sin duplicados.")
    print("Tamaño del dataset combinado:", df_combinado.shape)
    display(df_combinado.head())

else:
    print("No se encontró la columna 'producto' en ambos datasets para realizar la unión.")
```

Los registros con id_venta duplicados nos impiden agrupar directamente por producto, debemos antes agrupar por [id_venta, producto] quedarnos con el primer valor_venta y hacer la suma de los costos de marketing.
<BR>
Luego si agrupamos por producto, para saber el monto total de venta para cada uno.

```python
# Entonces grupamos primero por id_venta y producto
# de esa particion, nos quedamos con la primera venta y la suma de los costos de marketing
ventas_con_marketing_agg1 = ventas_con_marketing.groupby(["id_venta","producto"], as_index=False).agg(
    valor_venta=("valor_venta", "first"),
    costo_agg=("costo", "sum")
)

print("Cantidad de registros con id_venta únicos: ", len(ventas_con_marketing_agg1))
```

```python
# Verificamos la consistencia de los montos, ventas y costos de marketing
ventas_con_marketing_agg1[["id_venta","producto", "valor_venta", "costo_agg"]].sort_values(by=["id_venta"], ascending=True).head(6)
```

```python
# En la segunda agregacion, agrupamos solo por producto,
ventas_con_marketing_agg2 = ventas_con_marketing_agg1.groupby(["producto"], as_index=False).agg(
    ventas_producto_cm=("valor_venta", "sum"),
    costo_marketing=("costo_agg", "sum")
)

# Agregamos otra columna con la ganancia
ventas_con_marketing_agg2["ganacia_producto_cm"]= ventas_con_marketing_agg2["ventas_producto_cm"]-ventas_con_marketing_agg2["costo_marketing"]

# Finalmente, ordenamos y guardamos en el dataframe final
ventas_con_marketing_final = ventas_con_marketing_agg2.sort_values(by=["ganacia_producto_cm"], ascending=False)
```

```python
# Repetimos la agrupación pero ahora por mes,

ventas_con_marketing["mes"] = ventas_con_marketing["fecha_venta"].dt.month
ventas_con_marketing = ventas_con_marketing.sort_values("id_venta")

ventas_con_marketing_agg3 = ventas_con_marketing.groupby(["id_venta","mes"], as_index=False).agg(
    valor_venta=("valor_venta", "first"),
    costo_agg=("costo", "sum")
)

ventas_con_marketing_agg4 = ventas_con_marketing_agg3.groupby(["mes"], as_index=False).agg(
    ventas_producto_cm=("valor_venta", "sum"),
    costo_marketing=("costo_agg", "sum")
)

# Agregamos otra columna con la ganancia
ventas_con_marketing_agg4["ganacia_producto_cm"]= ventas_con_marketing_agg4["ventas_producto_cm"]-ventas_con_marketing_agg4["costo_marketing"]

# Finalmente, ordenamos y guardamos en el dataframe final
ventas_con_marketing_mes = ventas_con_marketing_agg4.sort_values(by=["mes"], ascending=True)
```

### Impacto marketing en ventas III

Una vez que tuve separados los valores de ventas con campaña y sin campaña para cada mes, el siguiente paso fue unir ambos resultados en un solo cuadro. Esto permite comparar mes a mes cuánto se vendió en cada modalidad y, sobre todo, visualizar si la presencia de campañas coincide con picos o variaciones relevantes.

Después de hacer el merge, transformé el dataframe al formato “largo”. Este formato es más práctico al momento de graficar, ya que deja cada combinación mes–modalidad en filas independientes y unifica los valores en una sola columna. De esta manera, los gráficos pueden diferenciar fácilmente entre ventas con marketing y ventas sin marketing, mostrando su evolución a lo largo del año.

```python
ventas_con_sin_marketing_mes = pd.merge(ventas_sin_marketing_mes, ventas_con_marketing_mes, on="mes", how="left")
ventas_con_sin_marketing_mes = ventas_con_sin_marketing_mes.sort_values(by=["ventas_producto_cm"], ascending=False)
ventas_con_sin_marketing_mes.columns
```

```python
# Transformar a formato largo
df_long = ventas_con_sin_marketing_mes[["mes","ventas_producto_cm", "ventas_producto_sm"]]
df_long = df_long.melt(id_vars="mes",
                        var_name="modalidad",
                        value_name="ventas")
df_long.head()
```

```python
# Limpieza estética
df_long["modalidad"] = df_long["modalidad"].replace({
    "ventas_producto_sm": "sin campaña",
    "ventas_producto_cm": "con campaña"
})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# --- Estilo general ---
sns.set_theme(style="whitegrid", context="talk")

# Paleta reducida
palette = sns.color_palette("Set2", n_colors=2)

plt.figure(figsize=(12, 6))

# --- Gráfico ---
sns.lineplot(
    data=df_long,
    x="mes",
    y="ventas",
    hue="modalidad",
    palette=palette,
    linewidth=2.8,
    marker="o",
    markersize=7,
)

# --- Títulos y etiquetas ---
plt.title("Evolución mensual según modalidad", fontsize=18, pad=20)
plt.xlabel("Mes", fontsize=14)
plt.ylabel("Ventas", fontsize=14)

# --- Ejes ---
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# --- Leyenda ---
plt.legend(
    title="Modalidad",
    title_fontsize=13,
    fontsize=11,
    loc="upper left",
    frameon=True,
    borderpad=1.1,
)

plt.tight_layout()
plt.show()

```
